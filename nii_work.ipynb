```python
import numpy as np
import nibabel as nib
import matplotlib.pyplot as plt
import scipy.ndimage as ndi
from google.colab import files
```

```python
img = nib.load("/content/sample1.nii")
```

```python
img
```

```python
data = img.get_fdata()
```

```python
data.shape
```

```python
def plotscan(data,x,plane):
  if plane == 'saggital':
    plt.imshow(ndi.rotate(np.fliplr(data[x,:,:]),-90),cmap='bone')
    plt.axis('off')
    plt.show()
  if plane == 'coronal':
    plt.imshow(ndi.rotate(np.fliplr(data[:,x,:]),-90),cmap='bone')
    plt.axis('off')
    plt.show()
  if plane == 'transverse':
    plt.imshow(ndi.rotate(np.fliplr(data[:,:,x]),90),cmap='bone')
    plt.axis('off')
    plt.show()
```

```python
for i in range(0,256,60):
  print(f'Slice {i}')
  plotscan(data,i,'saggital')
```

```python
for i in range(0,256,60):
  print(f'Slice {i}')
  plotscan(data,i,'coronal')
```

```python
for i in range(0,256,60):
  print(f'Slice {i}')
  plotscan(data,i,'transverse')
```
